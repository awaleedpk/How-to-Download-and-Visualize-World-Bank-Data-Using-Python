```python
import wbdata
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Set up the indicators we want
indicators = {"NY.GDP.MKTP.CD": "GDP (current US$)"}

# Set up the countries we want
countries = ["USA", "CHN", "JPN"]
```

```python
# Fetch the data
df = wbdata.get_dataframe(indicators, country=countries)
```

```python
# Clean and prepare the data
df = df.reset_index()
df['date'] = pd.to_datetime(df['date'])
df = df[(df['date'] >= '2000-01-01') & (df['date'] <= '2022-01-01')]
df = df.pivot(index='date', columns='country', values='GDP (current US$)')
```

```python
# Display the data
print(df)
```

```python
# Plot the data
df.plot(figsize=(12, 6))
plt.title("GDP Comparison")
plt.ylabel("GDP (current US$)")
plt.xlabel("Year")
plt.legend(title="Country")
plt.show()
```
